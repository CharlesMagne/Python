````python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1, "bo-", linewidth=2, markersize=12, label="First")
plt.plot(x,y2, "g^-", linewidth=2, markersize=12, label="Second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
#pyplot supports LaTeX!!
# plt.axis({xmin, xmax, ymin, ymax})
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("plot1.jpg")
````

Sometimes you need to make the axes logarithmic. For any given point to be plotted, its x and/or y-coordinate are transformed using the log function. Default base 10. <br/>
* `semilogx()` plots the x-axes on a log scale and the y in the original scale
* `semilogy()` plots the y-axes on the log scale and the x in the original scale
* `loglog()` plots both x and y on logarithmic scales. 
<br/>
$y=x^\alpha$ <br/>
$log(y)=log(x)^\alpha$ <br/>
$log(y) = (\alpha) log(x)$ <br/>
$y^1 = (\alpha)x^1$ <br/>
$\alpha$ is the angle of the line. Functions of the form y is equal to x to the power alpha are a stright lines on a `log()log()` plot

````python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x,y1, "bo-", linewidth=2, markersize=12, label="First")
plt.loglog(x,y2, "g^-", linewidth=2, markersize=12, label="Second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
# plt.axis({xmin, xmax, ymin, ymax})
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("plot2.jpg")
````

To generate numbers that are evenly spaces `np.linspace` on the logrithmic scale

````python
x = np.logspace(-1, 1, 40)
#logspace
# -1 = 0.1
# 10 = 1 end point 
# width 40
y1 = x**2.0
y2 = x**1.5
plt.loglog(x,y1, "bo-", linewidth=2, markersize=12, label="First")
plt.loglog(x,y2, "g^-", linewidth=2, markersize=12, label="Second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
# plt.axis({xmin, xmax, ymin, ymax})
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("plot3.jpg")
#all of the points along the axes aere evenly spaced
````

### Histograms
`hist` function from plt library. <br/>
First generate some random numbers using the `np.random` normal function to generate 1,000 samples from the standard normal distribution where the mean is equal to 0 and variance is equal to 1.

````python
x = np.random.normal(size=1000)
plt.hist(x);
````

https://bespokeblog.wordpress.com/2011/07/11/basic-data-plotting-with-matplotlib-part-3-histograms/ <br/>

Normalize the histogram and provide locations of bin edges used to construct the histogram. Default = 10 evenly spaces bins. Sometimes, need to specify the bins. 

````python
plt.hist(x, normed= True, bins=np.linspace(-5,5,21));
````

**gamma distribution** continuous probability density function that starts at 0 and goes all the way to positive infinity. Has two parameters. <br/><br/>
Draw large number of samples <br/>
Use `subplot()` can have several subplots within each figure. Takes three arguments, first two specify the number of rows and columns in the subplot and third is the plot number. 2 rows 3 columns = 6 subplots.

1 | 2 | 3
--| -- | --
4 | 5 | 6

````python
xa = np.random.gamma(2, 3, 100000)
````

````python
plt.figure()
plt.subplot(221)
ax1 = plt.hist(xa, bins = 30);
plt.subplot(222)
plt.hist(xa, bins=30, normed=True);
plt.subplot(223)
plt.hist(xa, bins=30, cumulative=True);
plt.subplot(224)
plt.hist(x, bins = 30, cumulative=True, normed=True, histtype="step");
````
